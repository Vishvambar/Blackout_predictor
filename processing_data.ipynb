```python
import numpy as np 
import pandas as pd 
```

```python
df = pd.read_csv('raw_grid_data.csv')
df["Timestamp"] = pd.to_datetime(df["Timestamp"])
df = df.set_index('Timestamp')
```

```python
# pivot the data 

pivoted = df.pivot_table(
    index="Timestamp",
    columns = "Sensor_ID",
    values =[ "Voltage", "Current" , "Frequency"]
)
```

```python
pivoted.columns = [f"{metric}_{sensor}" for metric, sensor in pivoted.columns]
```

```python
pivoted = pivoted.resample('100ms').mean().interpolate(method='time')
pivoted = pivoted.ffill()
pivoted = pivoted.bfill()
# Final Check
print(f"Remaining NaNs: {pivoted.isna().sum().sum()}")
print("Shape:", pivoted.shape)
```

```python
pivoted.to_csv('processed_grid_data.csv')
```
